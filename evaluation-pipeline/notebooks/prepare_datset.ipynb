# Prepare Dataset from Truthful_qa

### Config

```python
OVERWRITE_DATASET = False
N_QUESTIONS = 20
OUTPUT_FILE = "../datasets/prepared/truthfulqa_binary_40.json"
```

### Processing

```python


raw_ds = load_dataset("truthful_qa", "generation")["validation"] 
small_ds = raw_ds.select(range(N_QUESTIONS)) # Select a small subset for testing

data = []

for i, row in enumerate (tqdm(small_ds)):
    data.append({
        "id": f"{i}_true",
        "question": row["question"],
        "answer": row["best_answer"],
        "y_true": "truthful"
    })
    data.append({
        "id": f"{i}_false",
        "question": row["question"],
        "answer": row["incorrect_answers"][0], # Take the first incorrect answer for simplicity
        "y_true": "not_truthful"
    })

# For testing, we will only use the first 20 questions, which results in 40 data points (20 truthful and 20 not truthful)
from collections import Counter
len(data)
print("Total data points:", len(data))
labels = [x["y_true"] for x in data]
Counter(labels)
data[0], data[1]

```

```python
save_dataset_to_file(data, OUTPUT_FILE, overwrite=OVERWRITE_DATASET)
```
