# Prepare Datasets
### This notebook prepares datasets fro LLM-as-a-judge experiments

# Setup

```python
from pathlib import Path
from evaluation_pipeline.dataset import save_dataset_to_file, prepare_beavertails, prepare_truthfulqa, preview_dataset
```

# Config

```python
DATASET_NAME = "beaver_tails"
#options:
# "truthful_qa"
# "baever_tails"
N_SAMPLES = 100
OVERWRITE_DATASET = False

OUTPUT_DIR = Path("../datasets/prepared")
OUTPUT_DIR.mkdir(parents=True, exist_ok =True)

OUTPUT_FILE =OUTPUT_DIR / f"{DATASET_NAME}_binary_{N_SAMPLES}.json"
```

### Prepare dataset

```python

if DATASET_NAME == "truthful_qa":
    data = prepare_truthfulqa(N_SAMPLES)
elif DATASET_NAME == "beaver_tails":
    data = prepare_beavertails(N_SAMPLES)
else:
    raise ValueError(f"Unknown DATASET_NAME: {DATASET_NAME}")

preview_dataset(data)
```

# Save Dataset to the file

```python
save_dataset_to_file(data, OUTPUT_FILE, overwrite=OVERWRITE_DATASET)
```
