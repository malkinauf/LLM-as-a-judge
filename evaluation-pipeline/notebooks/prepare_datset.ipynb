# Prepare Datasets
### This notebook prepares datasets fro LLM-as-a-judge experiments

# Setup

```python
from evaluation_pipeline.dataset import save_dataset_to_file, prepare_beavertails, prepare_truthfulqa, preview_dataset, prepare_harmbench
from pathlib import Path
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s:%(name)s:%(message)s",
    force=True
)
```

# Config

```python
PROJECT_ROOT = Path.cwd().parent

DATASET_NAME = "harmbench"
#options: beavertails, truthfulqa, harmbench

N_SAMPLES = 100
OVERWRITE_DATASET = False
OUTPUT_DIR = PROJECT_ROOT / "datasets" / "prepared"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

OUTPUT_FILE = OUTPUT_DIR / f"{DATASET_NAME}_{N_SAMPLES}.json"
```

# Prepare Dataset 


```python
if DATASET_NAME not in {"beavertails", "truthfulqa", "harmbench"}:
    raise ValueError(
        "DATASET_NAME must be 'beavertails' or 'truthfulqa'"
    )
if DATASET_NAME == "beavertails":
    dataset = prepare_beavertails(n_samples=N_SAMPLES)
elif DATASET_NAME == "truthfulqa":
    dataset = prepare_truthfulqa(n_samples=N_SAMPLES)
elif DATASET_NAME == "harmbench":
    dataset = dataset = prepare_harmbench(n_samples=N_SAMPLES, include_disputed=False)
else:
    raise ValueError(f"Unknow dataset name: {DATASET_NAME}")
    
```

# Preview

```python
preview_dataset(dataset)
```

# Save dataset to the file

```python
save_dataset_to_file(dataset=dataset, path=OUTPUT_FILE, overwrite=OVERWRITE_DATASET)
```
